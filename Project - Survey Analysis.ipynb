```python
# Loading the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the data into a pandas DataFrame

df = pd. read_csv(r"C:\Users\PC User\Desktop\Me\DATA ANALYSIS\PROJECTS\FINISHED\Flight Satisfaction Survey\Flight Satisfaction Survey.csv")
df
```

```python
# Dropping unwanted columns

df.drop('Unnamed: 0', axis =1, inplace = True)
df.head()
```

```python
# Getting a summary of the dataset

df.info()
```

```python
# Getting basic statistics of numerical columns

df.describe()
```

```python
# Checking for missing values

df.isnull().sum()
```

```python
# Handling the missing values

df['Arrival Delay in Minutes'].fillna(df['Departure Delay in Minutes'], inplace=True)
df.isnull().sum()
```

```python
# Checking for duplicate values

df.loc[df.duplicated()]
```

```python
# 1. Demographic Analysis

    # Gender Distribution

plt.figure(figsize=(10, 8))
gender_counts = df['Gender'].value_counts()
colors = ['skyblue' if gender == 'Male' else 'salmon' for gender in gender_counts.index]
plt.pie(gender_counts, labels=gender_counts.index, autopct = '%1.2f%%', startangle=90, colors = colors)
plt.title('Gender Distribution')
plt.show()
```

```python
    # Age Distribution

plt.figure(figsize=(10, 8))
sns.histplot(x = 'Age', data = df, bins = 20, kde = True, hue = 'Gender', alpha = 0.6)
plt.grid(True)
plt.title('Age Distribution with Gender')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 8))
sns.histplot(x = 'Age', data = df, bins = 20, kde = True, hue = ('Customer Type'), alpha = 0.7)
plt.grid(True)
plt.title('Age Distribution with Customer Type')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Customer type Analysis

    # Customer type Distribution

plt.figure(figsize=(10, 8))
customer_type_counts = df['Customer Type'].value_counts()
plt.pie(customer_type_counts, labels=customer_type_counts.index, autopct = '%1.2f%%',startangle = 140)
plt.title('Customer Type Distribution')
plt.show()
```

```python
# 3. Travel and Class Analysis

    # Type of Travel Distribution

travel_counts= df['Type of Travel'].value_counts()

plt.figure(figsize=(10, 8))
sns.barplot(x=travel_counts.values, y=travel_counts.index)
for i, v in enumerate(travel_counts.values):
    plt.text(v + 0.1, i, str(v), ha='center', va='bottom')

plt.title('Type of Travel Distribution')
plt.xlabel('Count')
plt.ylabel('Type of Travel')
plt.show()
```

```python
    # Class Distribution

plt.figure(figsize=(10, 8))
class_counts = df['Class'].value_counts().sort_values()
sns.countplot(x = 'Class', data= df, order = class_counts.index, hue = 'Gender')
plt.title('Class Distribution with Gender')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 8))
class_counts = df['Class'].value_counts().sort_values()
sns.countplot(x = 'Class', data= df, order = class_counts.index, hue = 'Customer Type')
plt.title('Class Distribution with Customer Type')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# 4. Flight Distance Analysis

    # Flight Distance Distribution

plt.figure(figsize=(10, 8))
sns.histplot(x = 'Flight Distance', data = df, bins = 50, kde = True, hue='Gender', alpha = 0.7)
grid = True
plt.title('Flight Distance Distribution')
plt.xlabel('Distance (km)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 5. Service Quality and Satisfaction

    # Correlation of Satisfaction Levels

df_corr = df[['Inflight wifi service', 'Departure/Arrival time convenient', 'Ease of Online booking', 
                   'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort', 
                   'Inflight entertainment', 'On-board service', 'Leg room service', 
                   'Baggage handling', 'Check-in service', 'Inflight service', 'Cleanliness']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(df_corr, annot = True,linewidths=0.5)
plt.title('Correlation Matrix of Satisfaction levels')
plt.show()
```

```python
    # Services with their Ratings
```

```python
parameters = ['Inflight wifi service','Departure/Arrival time convenient', 'Ease of Online booking',
    'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
    'Inflight entertainment', 'On-board service', 'Leg room service',
    'Baggage handling', 'Check-in service', 'Inflight service',
    'Cleanliness']

plt.figure(figsize=(10, 20))
for i, param in enumerate(parameters):
    plt.subplot(5, 3, i+1)
    satisfaction_counts = df[param].value_counts()
    plt.pie(satisfaction_counts, labels=satisfaction_counts.index, autopct='%1.2f%%', startangle=140)
    plt.title(param)

plt.tight_layout()
plt.show()
```

```python
parameters = ['Inflight wifi service','Departure/Arrival time convenient', 'Ease of Online booking',
    'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
    'Inflight entertainment', 'On-board service', 'Leg room service',
    'Baggage handling', 'Check-in service', 'Inflight service',
    'Cleanliness']
service_means = df[parameters].mean().sort_values()

plt.figure(figsize = (10,8))
service_means.plot(kind = 'barh')
plt.xlabel('Average Rating')
plt.title('Average Ratings of the Services')
for index, value in enumerate(service_means):
    plt.text(value, index, f'{value:.2f}', va='center')

plt.tight_layout()
plt.show()
```

```python
# 6. Delay Analysis

    # Delay Distribution with Satisfaction levels

plt.figure(figsize=(10, 8))
sns.scatterplot(x = 'Departure Delay in Minutes', y = 'Arrival Delay in Minutes', data = df, hue = 'satisfaction')
plt.title('Relationship between Delay and Satisfaction ')
plt.xlabel('Departure Delay in Minutes')
plt.ylabel('Arrival Delay in Minutes')
plt.grid(True)
plt.legend(title = 'Satisfaction')
plt.show()
```

```python
    # Flight Distance with Arrival Delay

plt.figure(figsize=(10, 8))
sns.scatterplot(x = 'Flight Distance', y = 'Arrival Delay in Minutes', data = df, hue = 'Gender')
plt.title('Relationship between Flight Distance and Arrival Delay ')
plt.xlabel('Flight Distance in kms')
plt.ylabel('Arrival Delay in Minutes')
plt.grid(True)
plt.legend(title = 'Gender')
plt.show()
```

```python
    ## Overall Satisfaction Levels

plt.figure(figsize=(10, 8))
satisfaction_counts = df['satisfaction'].value_counts()
plt.pie(satisfaction_counts, labels = satisfaction_counts.index, autopct = '%1.2f%%', startangle = 90)
plt.title('Overall Satisfaction Levels')
plt.show()
```
